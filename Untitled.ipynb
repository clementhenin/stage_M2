```python
import pandas as pd
import pylab as plt
import numpy as np 
from sklearn import linear_model, preprocessing
from sklearn import grid_search
```

```python
def plot_correlations(data, min_match=0, size_tuple=(15,15)):
    """Plot the correlations between every variable of the data
    
    Input : pandas data frame with code, year as index and sorted
    in this order
    
    Output : graph of correlations
    """
    correlations = data.corr(min_periods=min_match)
    fig = plt.figure(figsize=size_tuple)
    norm = plt.cm.colors.Normalize(vmax=1, vmin=-1)
    plt.imshow(correlations, interpolation='nearest', norm=norm)
    plt.xticks(range(0, len(data.columns)), data.columns, rotation='vertical')
    plt.yticks(range(0, len(data.columns)), data.columns, rotation='horizontal')
    plt.colorbar()
```

```python
data = pd.read_csv('data_source/all_data.csv', index_col=['code', 'year'])
data = data[(data['savings_IMF'].apply(np.absolute) < 200).values + (data['savings_IMF'].isnull()).values]
len(data.index.levels[0])
```

```python
plot_correlations(data)
```

# Creating explicative variables from diverse data sources

```python
rich_share_var = ["D1_WB", "QU1_WB", "D9_WIID", "Q5_WIID", "D1_WID", "V1_WID", "P1_WID", 
"top_0.5_income_share_WID", "Pr1_WID", "top_0.05_income_share_WID", "top_0.01_income_share_WID", 
"P90/P10_OECD", "P90/P50_OECD", "P50/P10_OECD", "S80/S20_OECD", "S90/S10_OECD"]
for var in rich_share_var:
    data.loc[:, var] = (data.loc[:, var] - data.loc[:, var].mean()) / data.loc[:, var].std()
plot_correlations(data[rich_share_var], size_tuple=(10,10))
```

```python
GDP_PC_var = ["GDP_PC_WB", "GDP_PC_PWT"]
print "Variables caracteristics : "
for var in GDP_PC_var:
    print "mean : ", data[var].mean(), "\nStd : ", data[var].std()
plot_correlations(data[GDP_PC_var], size_tuple=(4,4))
```

```python
savings = ["savings_WB", "savings_OECD", "savings_IMF"]
for var in savings:
    print "mean : ", data[var].mean(), "\nStd : ", data[var].std()
plot_correlations(data[savings], size_tuple=(4,4))
print data[savings].corr()
```

```python
investments = ["investments_WB", "invest_IMF"]
for var in investments:
    print "mean : ", data[var].mean(), "\nStd : ", data[var].std()
plot_correlations(data[investments], size_tuple=(4,4))
print data[investments].corr()
```

```python
all_variable = ['GDP_growth_WB', "openness_WB", "investments", "years_schooling", "savings", 'gini', 
                "price_level_investment_PWT", "gov_consumption_WB", "inflation_WB", "fertility_WB", 
                'rich_share', 'GDP_PC']

data['rich_share'] = data[rich_share_var].mean(axis=1)
data['GDP_PC'] = data[GDP_PC_var].mean(axis=1)
data['savings'] = data[savings].mean(axis=1)
data['investments'] = data[investments].mean(axis=1)
data['gini'] = data[ginis].mean(axis=1)
```

```python
dataset = data[all_variable]
dataset = dataset.dropna(how='all')
```

```python
dataset.reset_index(inplace=True)
dataset.sort_values(['code', 'year'], inplace=True)
dataset.set_index(['code'], inplace=True)
new_frame = pd.DataFrame()
for country in set(dataset.index.values):
    sel = dataset.loc[country]
    if sel.shape != (len(all_variable) + 1,):
        try:
            sel.loc[:, 'years_schooling'] = sel['years_schooling'].interpolate()
        except TypeError:
            None
        new_frame = pd.concat([new_frame, sel])
dataset = new_frame.reset_index().set_index(['code', 'year'])
```

# Linear regression

```python
X = dataset[dataset.keys().tolist()[1:]]
Y = dataset['GDP_growth_WB']
```

```python
clf = linear_model.LinearRegression()
clf.fit(X, Y)

x = np.linspace(0, 209, len(dataset))
plt.figure(figsize=(15,7))
plt.plot(x, Y, 'o', label="observed growth")
plt.plot(x, clf.predict(X), 'r--.', label="'predicted'")
plt.legend()
plt.ylabel("Growth")
plt.xlabel("Country/period")

print "Original variance of the sample : ", Y.var()
print "Variance of the residu :", (Y - clf.predict(X)).var()
print "Explained variance share : ", (Y.var() - (Y - clf.predict(X)).var()) / Y.var()
```

### Grid search cross validation

```python
parameters = {'alpha':np.logspace(-5, 0, 50)}
lm = linear_model.Lasso()
clf = grid_search.RandomizedSearchCV(lm, parameters, cv=3)
clf.fit(X, Y)
print "Best alpha parameter : ", clf.best_params_
print "Best score : ", clf.best_score_

x = np.linspace(0, 209, len(dataset))
plt.figure(figsize=(15,7))
plt.plot(x, Y, 'o', label="observed growth")
plt.plot(x, clf.predict(X), 'r--.', label="'predicted'")
plt.legend()
plt.ylabel("Growth")
plt.xlabel("Country/period")

print "Original variance of the sample : ", Y.var()
print "Variance of the residu :", (Y - clf.predict(X)).var()
print "Explained variance share : ", (Y.var() - (Y - clf.predict(X)).var()) / Y.var()
```

# Same thing after resampling the data to 5 years periods

```python
def resample(df, period):
    """Resample the frame df to period (in year)

    Input :
    -df : dataframe pandas with no index the time dimension exist as a year
        columns (column of integer)
    - period : integer : new wanted period

    Output : resampled frame
    """
    period = float(period)
    tuples = zip(df['code'],
                 (((df['year'].values) // period) * period) + period / 2.)
    df_copy = df.copy()
    df_copy.index = pd.MultiIndex.from_tuples(tuples)
    df_copy = df_copy.groupby(level=[0, 1]).mean().dropna(how='all')
    del df_copy['year']
    df_copy.index.names = ['code', 'year']
    return df_copy.sort_index()
```

```python
X_res = resample(X.reset_index(), 5)
Y_res = resample(Y.reset_index(), 5).values
```
