```python
import pandas as pd
import datetime
```

```python
data = pd.read_csv("barro_data.csv")
country_list = pd.read_csv("../data_source/country_code_list.csv")
data.head()
```

We only data keep posterior to 1960. We also remove the countries that do not appear in the list of countries. 

```python
data = data[data['year']>=1960.]
data = data[data['code'].map(lambda x: x in country_list['code'].values)]
data.info()
```

# Resampling the data

```python
period = "5AS"
data.index = data['year'].map(lambda x: datetime.datetime(int(x), 1, 1, 0, 0, 0))
re_data = pd.DataFrame()
for country in set(data['code']):
    sel = data[data['code']==country].resample(period)
    sel['code'] = [country] * len(sel)
    re_data = pd.concat([re_data, sel], axis=0)
```
